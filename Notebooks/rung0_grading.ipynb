## Compute Goodness, Precision, Accuracy

The Goodness, Precision, and Accuracy will be computed for the overall dataset as well as for individual lenses. This grading scheme is identical to that of [Ding et al. 2021](https://academic.oup.com/mnras/article/503/1/1096/6146088), section 2.7.

$$\chi^2 = \frac{1}{N}\sum_i \chi_i^2$$
$$\chi_i^2 = \left(\frac{x_{i, obs} - x_i}{\sigma_i}\right)^2$$

```python
chi_i_squared = ((x_i_obs - x_i) / sigma_i)**2
chi_squared = (1 / N) * np.sum(chi_i_squared)
print(chi_squared)
```

$$P = \frac{1}{N}\sum_i P_i$$
$$P_i = \frac{\sigma_i}{x_i}$$

```python
P_i = sigma_i / x_i
P = (1 / N) * np.sum(P_i)
print(P)
```

$$A_i = \frac{x_{i, obs} - x_i}{x_i}$$
$$A = \frac{1}{N}\sum_i A_i$$

```python
A_i = (x_i_obs - x_i) / x_i
A = (1 / N) * np.sum(A_i)
print(A)
```
