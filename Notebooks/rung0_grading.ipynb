```python
import pandas as pd
import numpy as np

def get_csv_header_pandas(filepath):
    # Pandas automatically treats the first row as the header by default
    df = pd.read_csv(filepath) 
    # The .columns attribute holds the header names
    header_list = df.columns.tolist()
    return header_list
```

## Compute Goodness, Precision, Accuracy

$$\chi^2 = \frac{1}{N}\sum_i \left(\frac{x_{i, obs} - x_i}{\sigma_i}\right)^2$$

```python
chi_squared = (1 / N) * np.sum(((x_i_obs - x_i) / sigma_i)**2)
print(chi_squared)
```

$$P = \frac{1}{N}\sum_i \frac{\sigma_i}{x_i}

```python
P = (1 / N) * np.sum(sigma_i / x_i)
print(P)
```

$$A = \frac{1}{N}\sum_i \frac{x_{i, obs} - x_i}{x_i}

```python
A = (1 / N) * np.sum((x_i_obs - x_i) / x_i)
print(A)
```
